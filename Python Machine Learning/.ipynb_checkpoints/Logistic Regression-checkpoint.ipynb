```python
def sigmoid(z):
    sigVec = np.mat(1/(1+np.exp(-z)))
    return sigVec
```

```python
def predict(weight, test, testLabel):
    test = np.mat(test)
    testLabel = np.mat(testLabel)
    m,n = test.shape
    prediction = sigmoid(test*weight)
    print(prediction.transpose())
    for i in range(m):
        if prediction[i]<0.5:
            prediction[i]=0
        else:
            prediction[i]=1
    return prediction
```
